```python
# Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance is ${self.balance}.")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance is ${self.balance}.")
        else:
            print("Insufficient funds or invalid amount.")

# Example usage:
account = BankAccount("123456", "John Doe", 1000)
account.deposit(500)
account.withdraw(200)




# Class Definition: The BankAccount class represents a bank account with attributes like account number, account holder name, and balance. It includes methods to deposit and withdraw money.

# Attributes:
# account_number: Unique identifier for the account.
# account_holder_name: Name of the account holder.
# balance: Current balance of the account.

# Methods:
# deposit(amount): Adds a specified amount to the balance.
# withdraw(amount): Subtracts a specified amount from the balance, provided there are sufficient funds.

# Constructor (__init__): Initializes the account with an account number, holder's name, and an optional initial balance.

# Deposit Method: Adds money to the account, checks for positive amount.

# Withdraw Method: Removes money from the account if sufficient funds are available.
```

```python
# Problem 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_bonus(self, percentage):
        return self.salary * 12 * (percentage / 100)

    def display_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Monthly Salary: ${self.salary}")

# Example usage:
employee = Employee("E123", "Alice Smith", 5000)
print(f"Yearly Bonus (10%): ${employee.calculate_bonus(10)}")
employee.display_details()


# Class Definition: The Employee class represents an employee with attributes like employee ID, name, and salary. It includes methods to calculate the yearly bonus and display employee details.

# Attributes:
# employee_id: Unique identifier for the employee.
# name: Name of the employee.
# salary: Monthly salary of the employee.

# Methods:
# calculate_bonus(): Calculates the yearly bonus (typically a percentage of the salary).
# display_details(): Displays employee details including ID, name, and salary.

# Constructor (__init__): Initializes the employee with an ID, name, and salary.
# Calculate Bonus Method: Computes the yearly bonus based on a percentage.
# Display Details Method: Prints out the employee's details.
```

```python
# Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

class VehicleRental:
    def __init__(self):
        self.vehicles = {'V001': True, 'V002': True, 'V003': False}  # True means available

    def rent_vehicle(self, vehicle_id):
        if self.vehicles.get(vehicle_id, False):
            self.vehicles[vehicle_id] = False
            print(f"Vehicle {vehicle_id} has been rented.")
        else:
            print(f"Vehicle {vehicle_id} is not available.")

    def return_vehicle(self, vehicle_id):
        if vehicle_id in self.vehicles:
            self.vehicles[vehicle_id] = True
            print(f"Vehicle {vehicle_id} has been returned.")
        else:
            print(f"Vehicle {vehicle_id} does not exist.")

    def display_available_vehicles(self):
        available_vehicles = [vid for vid, available in self.vehicles.items() if available]
        print("Available Vehicles:", ', '.join(available_vehicles))

# Example usage:
rental_system = VehicleRental()
rental_system.display_available_vehicles()
rental_system.rent_vehicle('V001')
rental_system.return_vehicle('V003')
rental_system.display_available_vehicles()



# Class Definition: The VehicleRental class represents a vehicle rental system. It includes methods to rent, return, and display available vehicles.

# Attributes:
# vehicles: A dictionary where keys are vehicle IDs and values are their availability status.
# Methods:

# rent_vehicle(vehicle_id): Marks a vehicle as rented if it's available.
# return_vehicle(vehicle_id): Marks a vehicle as returned and available.
# display_available_vehicles(): Lists all available vehicles.

# Constructor (__init__): Initializes the system with a set of vehicles and their availability.

# Rent Vehicle Method: Changes the status of a vehicle to rented if it's available.

# Return Vehicle Method: Changes the status of a vehicle to available upon return.

# Display Available Vehicles Method: Lists vehicles that are currently available.
```

```python
# Problem 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def borrow(self):
        if self.available:
            self.available = False
            return True
        return False

    def return_book(self):
        self.available = True

class Library:
    def __init__(self):
        self.books = {}

    def add_book(self, book):
        self.books[book.title] = book

    def borrow_book(self, title):
        book = self.books.get(title)
        if book and book.borrow():
            print(f"Borrowed '{title}' successfully.")
        else:
            print(f"Book '{title}' is not available or does not exist.")

    def display_available_books(self):
        available_books = [title for title, book in self.books.items() if book.available]
        print("Available Books:", ', '.join(available_books))

# Example usage:
library = Library()
library.add_book(Book("1984", "George Orwell"))
library.add_book(Book("To Kill a Mockingbird", "Harper Lee"))

library.display_available_books()
library.borrow_book("1984")
library.display_available_books()



# Class Definitions:
# Book: Represents a book with attributes like title, author, and availability status.
# Library: Manages a collection of books and operations like adding, borrowing, and displaying books.
# Attributes:

# books: A dictionary where keys are book titles and values are Book objects.
# Methods:

# add_book(book): Adds a new book to the library.
# borrow_book(title): Marks a book as borrowed if it's available.
# display_available_books(): Lists all books that are available for borrowing.

# Book Class: Represents a book with its title, author, and availability status.
# Library Class: Manages books, including adding new ones, borrowing, and listing available books.
```

```python
# Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

class Product:
    def __init__(self, product_id, name):
        self.product_id = product_id
        self.name = name
        self.quantity = 0

    def update_quantity(self, quantity):
        self.quantity = quantity

class ProductInventory:
    def __init__(self):
        self.products = {}

    def add_product(self, product, quantity):
        product.update_quantity(quantity)
        self.products[product.product_id] = product

    def update_quantity(self, product_id, quantity):
        product = self.products.get(product_id)
        if product:
            product.update_quantity(quantity)
            print(f"Updated {product.name} to quantity {quantity}.")
        else:
            print(f"Product ID {product_id} does not exist.")

    def display_products(self):
        for product_id, product in self.products.items():
            print(f"Product ID: {product_id}, Name: {product.name}, Quantity: {product.quantity}")

# Example usage:
inventory = ProductInventory()
inventory.add_product(Product("P001", "Laptop"), 10)
inventory.add_product(Product("P002", "Smartphone"), 20)

inventory.display_products()
inventory.update_quantity("P001", 15)
inventory.display_products()


# Class Definition: The ProductInventory class manages a collection of products with methods to add, update quantity, and display products.

# Attributes:
# products: A dictionary where keys are product IDs and values are Product objects.

# Methods:
# add_product(product, quantity): Adds a new product with a specified quantity.
# update_quantity(product_id, quantity): Updates the quantity of a specified product.
# display_products(): Lists all products with their quantities.

# Product Class: Represents a product with its ID, name, and quantity.
# ProductInventory Class: Manages products, including adding new ones, updating quantities, and displaying all products.
```

```python
# Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        if self.width:
            return self.length * self.width
        else:
            return self.length ** 2  # Assuming square if width is not provided

    def calculate_perimeter(self):
        if self.width:
            return 2 * (self.length + self.width)
        else:
            return 4 * self.length  # Assuming square if width is not provided

# Example usage:
rectangle = Shape(10, 5)
print(f"Rectangle Area: {rectangle.calculate_area()}")
print(f"Rectangle Perimeter: {rectangle.calculate_perimeter()}")

square = Shape(4)
print(f"Square Area: {square.calculate_area()}")
print(f"Square Perimeter: {square.calculate_perimeter()}")



# Class Definition: The Shape class represents a shape with attributes like length, width, and height. It includes methods to calculate the area and perimeter of the shape.

# Attributes:
# length: The length of the shape.
# width: The width of the shape (for rectangles).
# height: The height of the shape (for 3D shapes).

# Methods:
# calculate_area(): Computes the area of the shape. (Note: This method assumes a rectangular shape for simplicity).
# calculate_perimeter(): Computes the perimeter of the shape.

# Constructor (__init__): Initializes the shape with length, and optionally width and height.
# Calculate Area Method: Computes the area. Assumes a square if only length is provided.
# Calculate Perimeter Method: Computes the perimeter. Assumes a square if only length is provided.
```

```python
# Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def calculate_average(self):
        return sum(self.grades) / len(self.grades) if self.grades else 0

    def display_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        print(f"Grades: {self.grades}")
        print(f"Average Grade: {self.calculate_average()}")

# Example usage:
student = Student("S123", "Emily Johnson", [85, 90, 78, 92])
student.display_details()


# Class Definition: The Student class represents a student with attributes like student ID, name, and grades. It includes methods to calculate the average grade and display student details.

# Attributes:
# student_id: Unique identifier for the student.
# name: Name of the student.
# grades: List of grades for the student.

# Methods:
# calculate_average(): Computes the average of the student's grades.
# display_details(): Displays the student's details including ID, name, and grades.

# Constructor (__init__): Initializes the student with ID, name, and grades.
# Calculate Average Method: Computes the average grade from the list of grades.
# Display Details Method: Prints the student's ID, name, grades, and average grade.
```

```python
# Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post):
        self.posts.append(post)
        print(f"Added post: {post}")

    def display_posts(self):
        print(f"Posts by {self.username}:")
        for post in self.posts:
            print(f"- {post}")

    def search_posts(self, keyword):
        results = [post for post in self.posts if keyword in post]
        print(f"Posts containing '{keyword}':")
        for result in results:
            print(f"- {result}")

# Example usage:
profile = SocialMediaProfile("john_doe")
profile.add_post("Just had a great day at the beach!")
profile.add_post("Looking forward to the weekend.")
profile.display_posts()
profile.search_posts("weekend")



# Class Definition: The SocialMediaProfile class represents a social media profile with attributes like username and posts. It includes methods to add posts, display posts, and search for posts by keyword.

# Attributes:
# username: Username of the profile.
# posts: List of posts made by the user.

# Methods:
# add_post(post): Adds a new post to the profile.
# display_posts(): Displays all posts made by the user.
# search_posts(keyword): Searches for posts containing a specific keyword.

# Constructor (__init__): Initializes the profile with a username and an empty list of posts.
# Add Post Method: Adds a new post to the list.
# Display Posts Method: Prints all posts made by the user.
# Search Posts Method: Finds and prints posts containing a specific keyword.
```

```python
# Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

class ToDoList:
    def __init__(self):
        self.tasks = {}

    def add_task(self, task, due_date=None):
        self.tasks[task] = due_date
        print(f"Added task: '{task}' with due date {due_date}")

    def mark_task_completed(self, task):
        if task in self.tasks:
            del self.tasks[task]
            print(f"Task '{task}' marked as completed.")
        else:
            print(f"Task '{task}' not found.")

    def display_pending_tasks(self):
        if not self.tasks:
            print("No pending tasks.")
        else:
            print("Pending Tasks:")
            for task, due_date in self.tasks.items():
                print(f"- {task} (Due: {due_date})")

# Example usage:
todo_list = ToDoList()
todo_list.add_task("Finish homework", "2024-09-05")
todo_list.add_task("Buy groceries")
todo_list.display_pending_tasks()
todo_list.mark_task_completed("Finish homework")
todo_list.display_pending_tasks()



# Class Definition: The ToDoList class represents a to-do list with attributes like tasks and due dates. It includes methods to add tasks, mark tasks as completed, and display pending tasks.

# Attributes:
# tasks: A dictionary where keys are task descriptions and values are due dates (or None if no due date).

# Methods:
# add_task(task, due_date=None): Adds a new task to the list with an optional due date.
# mark_task_completed(task): Marks a task as completed by removing it from the list.
# display_pending_tasks(): Displays all tasks that have not been marked as completed.

# Constructor (__init__): Initializes the to-do list with an empty dictionary for tasks.
# Add Task Method: Adds a task with an optional due date.
# Mark Task Completed Method: Removes a task from the list, marking it as completed.
# Display Pending Tasks Method: Lists all tasks that are still pending.
```
