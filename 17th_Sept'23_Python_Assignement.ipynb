```python
#16 Write a Python program to find the GCD (Greatest Common Divisor) of two numbers using a for loop.

# Python program to find the GCD (Greatest Common Divisor) of two numbers using a for loop

def gcd(a, b):
    for i in range(1, min(a, b) + 1):  # Iterates from 1 to the smallest number
        if a % i == 0 and b % i == 0:  # Checks if i is a divisor of both numbers
            gcd = i  # Updates the GCD

    return gcd

print("GCD of 36 and 60:", gcd(36, 60))


# Explanation:

# for i in range(1, min(a, b) + 1): Tests each number up to the smaller of the two numbers.
# if a % i == 0 and b % i == 0: Checks if i divides both numbers without a remainder.
```

```python
#22 Create a Python program that calculates the square root of a number using a for loop.

# Python program to calculate the square root of a number using a for loop

def square_root(number, iterations=10):
    guess = number
    for _ in range(iterations):
        guess = (guess + number / guess) / 2
    return guess

number = 25
sqrt = square_root(number)

print(f"Square root of {number} is approximately: {sqrt}")


# explanation:

# guess = (guess + number / guess) / 2: Updates the guess based on the Babylonian method.
# iterations: Number of iterations to refine the guess. More iterations yield a more accurate result.
```

```python
# Write a Python program to find the LCM (Least Common Multiple) of two numbers using a for loop.

# Python program to find the LCM (Least Common Multiple) of two numbers using a for loop

def lcm(a, b):
    max_num = max(a, b)
    for i in range(max_num, a * b + 1, max_num):
        if i % a == 0 and i % b == 0:
            return i
        

number1 = 12
number2 = 15
result = lcm(number1, number2)

print(f"LCM of {number1} and {number2} is:", result)


# Explanation:

# max_num = max(a, b): Starts checking from the larger of the two numbers.
# range(max_num, a * b + 1, max_num): Iterates through multiples of the larger number.
# if i % a == 0 and i % b == 0: Checks if the current number is divisible by both inputs, thus the LCM.
```

```python
# 1. Explain the purpose of the `map()` function in Python and provide an example of how it can be used to apply a function to each element of an iterable.

# Example of using map() to add 2 to each element in a list
def add_two(x):
    return x + 2

numbers = [1, 2, 3, 4]
result = list(map(add_two, numbers))
print(result)  # Output: [3, 4, 5, 6]


# Explanation:
# map(add_two, numbers): Applies the add_two function to each element of the numbers list.
# list(map(...)): Converts the map object to a list for easier viewing.
```

```python
# 2. Write a Python program that uses the `map()` function to square each element of a list of numbers.

# Python program to square each element of a list using map()

numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]


# Explanation:
# lambda x: x ** 2: An anonymous function that squares its input.
# map(lambda x: x ** 2, numbers): Applies the squaring function to each element of numbers
```

```python
# 3. How does the `map()` function differ from a list comprehension in Python, and when would you choose one over the other?

# map(): Ideal for applying a single function to each element of an iterable. It is more concise when using a predefined function and can handle multiple iterables simultaneously.
# List Comprehension: More flexible and often more readable when performing simple transformations. It allows for additional logic (e.g., conditionals) within the comprehension.

# Squaring elements using list comprehension
numbers = [1, 2, 3, 4, 5]
squared_numbers = [x ** 2 for x in numbers]
print(squared_numbers)  # Output: [1, 4, 9, 16, 25]


# When to Choose:
# map(): When you want to apply a single function across elements, especially if that function is already defined.
# List Comprehension: When you need additional logic or find it more readable for simple transformations.
```

```python
# 1. Explain the concept of recursion in Python. How does it differ from iteration?

# Base Case: A condition under which the recursion ends.
# Recursive Case: The part of the function where it calls itself with modified arguments to gradually approach the base case.

def factorial(n):
    if n == 1:  # Base case
        return 1
    return n * factorial(n - 1)  # Recursive case
```

```python
# 2. Write a Python program to calculate the factorial of a number using recursion.

def factorial(n):
    if n == 1:  # Base case
        return 1
    return n * factorial(n - 1)  # Recursive case

print(factorial(5))  

# Explanation:
# Base Case: When n == 1, return 1.
# Recursive Case: Multiply n by the factorial of n-1.
```

```python
# 3. Create a recursive Python function to find the nth Fibonacci number.

def fibonacci(n):
    if n == 0:  # Base case
        return 0
    elif n == 1:  # Base case
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)  # Recursive case

print(fibonacci(6)) 


# Explanation:
# Base Cases: n == 0 returns 0, n == 1 returns 1.
# Recursive Case: Sum of the two preceding Fibonacci numbers.
```

```python
# 4. Write a recursive Python function to calculate the sum of all elements in a list.

def sum_list(lst):
    if not lst:  # Base case: empty list
        return 0
    return lst[0] + sum_list(lst[1:])  # Recursive case

print(sum_list([1, 2, 3, 4]))  


# Explanation:
# Base Case: An empty list returns 0.
# Recursive Case: Sum the first element with the sum of the remaining list.
```

```python
# 5. How can you prevent a recursive function from running indefinitely, causing a stack overflow error?

# To prevent a recursive function from running indefinitely:

# Ensure the base case is reachable.
# Use recursion only when necessary and consider iterative solutions if possible.
# Python has a recursion limit (sys.setrecursionlimit()), but deep recursion should be avoided as it may cause a stack overflow.
```

```python
# 6. Create a recursive Python function to find the greatest common divisor (GCD) of two numbers using the Euclidean algorithm.

def gcd(a, b):
    if b == 0:  # Base case
        return a
    return gcd(b, a % b)  # Recursive case

print(gcd(48, 18)) 


# Explanation:
# Base Case: When b == 0, return a.
# Recursive Case: Call gcd with b and a % b.
```

```python
# 7. Write a recursive Python function to reverse a string.

def reverse_string(s):
    if len(s) == 0:  # Base case
        return ""
    return s[-1] + reverse_string(s[:-1])  # Recursive case

print(reverse_string("hello"))  


# Explanation:
# Base Case: An empty string returns an empty string.
# Recursive Case: Concatenate the last character with the reverse of the remaining string
```

```python
# 8. Create a recursive Python function to calculate the power of a number (x^n).

def power(x, n):
    if n == 0:  # Base case
        return 1
    return x * power(x, n - 1)  # Recursive case

print(power(2, 3)) 


# Explanation:
# Base Case: n == 0 returns 1.
# Recursive Case: Multiply x by power(x, n - 1)
```

```python
# 9. Write a recursive Python function to find all permutations of a given string.

def permutations(s):
    if len(s) == 1:  # Base case
        return [s]
    perms = []
    for i in range(len(s)):
        for perm in permutations(s[:i] + s[i+1:]):
            perms.append(s[i] + perm)
    return perms

print(permutations("abc"))  

# Explanation:
# Base Case: A single-character string returns a list containing that string.
# Recursive Case: Generate permutations by fixing each character and permuting the rest.
```

```python
# 10. Write a recursive Python function to check if a string is a palindrome.

def is_palindrome(s):
    if len(s) <= 1:  # Base case
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome(s[1:-1])  # Recursive case

print(is_palindrome("radar"))  


# Explanation:
# Base Case: A string of length 1 or 0 is a palindrome.
# Recursive Case: Compare the first and last characters and recursively check the rest.
```

```python
# 11. Create a recursive Python function to generate all possible combinations of a list of elements.

def combinations(lst):
    if len(lst) == 0:  # Base case
        return [[]]
    combs = []
    for c in combinations(lst[1:]):
        combs += [c, [lst[0]] + c]
    return combs

print(combinations([1, 2, 3]))  

# Explanation:
# Base Case: An empty list returns a list containing an empty list.
# Recursive Case: Include or exclude the first element and find combinations of the rest.
```

```python
# 2. Write a Python program that uses the `map()` function to apply a function to each element of a list

# Define a function to be applied
def square(x):
    return x ** 2

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Apply the function to each element using map
squared_numbers = list(map(square, numbers))

print(squared_numbers) 


# Explanation:
# The map() function applies the square function to each element in the numbers list.
# It returns an iterator that is converted to a list with list(), resulting in a list of squared numbers.
```

```python
# 3. How does the `filter()` function work in Python, and when would you use it?

# Define a function that returns True for even numbers
def is_even(x):
    return x % 2 == 0

# List of numbers
numbers = [1, 2, 3, 4, 5, 6]

# Filter even numbers using filter
even_numbers = list(filter(is_even, numbers))

print(even_numbers) 


# Explanation:
# The filter() function is useful for extracting specific elements from a collection based on a condition. It helps in cases where you want to process only certain elements that meet criteria defined by a function.
```

```python
# 4. Create a Python program that uses the `reduce()` function to find the product of all elements in a list.

from functools import reduce

# Define a function to multiply two numbers
def multiply(x, y):
    return x * y

# List of numbers
numbers = [1, 2, 3, 4, 5]

# Apply the function to all elements using reduce
product = reduce(multiply, numbers)

print(product) 


# Explanation:
# The reduce() function from the functools module applies the multiply function cumulatively to the elements of the numbers list, resulting in the product of all elements.
# It reduces the list to a single value by combining elements through the specified function.
```

```python
# 1. Explain the purpose of docstrings in Python functions and how to write them.

# Purpose of Docstrings:
# Documentation: Docstrings provide a convenient way to document what a function does, its parameters, and its return value. They serve as an in-code reference for users and developers.
# Help Functions: They allow tools like IDEs and documentation generators to automatically retrieve and display information about the function, enhancing code readability and maintainability.
# How to Write Docstrings:

# Syntax: Docstrings are written within triple quotes (""" or ''') immediately following the function definition.
# Content: A docstring should describe the function's purpose, parameters, return values, and any side effects or exceptions.

def add(x, y):
    """
    Add two numbers and return the result.

    Parameters:
    x (int or float): The first number.
    y (int or float): The second number.

    Returns:
    int or float: The sum of x and y.
    """
    return x + y

# Explanation:
# The docstring provides a clear description of what the function does, its input parameters, and its return value.
# It helps other developers understand the purpose and usage of the function without needing to read through the entire code.
```

```python
add(6,8)
```
